```python
import gym
import matplotlib.pyplot as plt
import numpy as np

env = gym.make("CartPole-v0")
env.reset()

i_arr = range(2500)
rewards = np.zeros(2500)
velocities = np.zeros(2500)
power = np.zeros(2500)

for i in range(2500):
    env.render()
    newState, rew, _, _ = env.step(env.action_space.sample())
    velocities[i] = np.sqrt(newState[7]**2 + newState[8]**2)
    # target velocity, with calculations described as above
    if (velocities[i] > 2.428371031919227):
        velocities[i] = 2.428371031919227
    power[i] = 2.448*(np.sum(np.abs(newState[9:])))
    if (power[i] > 50 or np.isnan(power[i])):
        power[i] = 50
    rewards[i] = rew

env.close()

cots = power/velocities

for i in range(2500):
    if (cots[i] > 50 or np.isnan(cots[i])):
        cots[i] = 50

print("Average cot")
print(np.average(cots))

plt.figure()
plt.plot(i_arr, cots)
plt.show()

print("Average power")
print(np.average(power))

plt.figure() 
plt.plot(i_arr, power)
plt.show()
```
